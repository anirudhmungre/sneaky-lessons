```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../Resources/residential-building.csv', header=None)
df
```

```python
X = df.drop([107], axis=1)
y = df[107]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
reg = LinearRegression().fit(X_train_scaled, y_train)
reg.score(X_test_scaled, y_test)
```

```python
# Select the columns that have a nonzero value from the LASSO regression.
reg = Lasso(max_iter=2000).fit(X_train_scaled, y_train)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.feature_selection import SelectFromModel
sel = SelectFromModel(reg)
sel.fit(X_train_scaled, y_train)
```

```python
X_selected_train, X_selected_test, y_train, y_test = train_test_split(sel.transform(X), y, random_state=1)
scaler = StandardScaler().fit(X_selected_train)
X_selected_train_scaled = scaler.transform(X_selected_train)
X_selected_test_scaled = scaler.transform(X_selected_test)
```

```python
reg = LinearRegression().fit(X_selected_train_scaled, y_train)
reg.score(X_selected_test_scaled, y_test)
```
