```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Load iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# The iris data
data = iris.data
data[:5]
```

```python
# Feature names
iris.feature_names
```

```python
# Label (iris species) of each sample
iris.target[:5]
```

```python
# Plot sepal length by petal length
plt.scatter(data[:,0], data[:,2])
plt.show()
```

```python
# Color samples by label
plt.scatter(data[:,0], data[:,2], c=iris.target)
plt.show()
```

```python
# Scale the data
scaler = StandardScaler()
features_standardized = scaler.fit_transform(data)
```

```python
# Instantiate the model
dbscan = DBSCAN(eps=0.6, min_samples=5)
dbscan.fit(features_standardized)
```

```python
# A sample of the DBSCAN cluster labels
labels = dbscan.labels_
labels[::10]
```

```python
# The model returns 2 clusters, as well as outliers (noise)
set(labels)
```

```python
# Plot the results, clustered by DBSCAN
plt.scatter(data[:,0], data[:,2], c=labels)
plt.show()
```
